```python
#Inheritence

#1)SINGLE LEVEL INHERITANCE 

class A:#SUPER CLASS/PARENT CLASS  #Grand father
    def feature1(self):
        print("1")
    def feature2(self):
        print("2")
        
class D:
    def feature10(self):
        print("10")
        
        
        
class B(A):#inheriting #SUBCLASS/CHILD CLASS #father
    def feature3(self):
        print("3")
    def feature4(self):
        print("4")
        
        
        
#2)MULTILEVEL INHERITANCE
class C(B):#child
    def feature5(self):
        print("5")


        
        
#3)MULTIPLE INHERITANCE 
class E(A,D):
    def feature20(self):
        print("20")
        

a1= A()

a1.feature1()
a1.feature2()

b1=B()
b1.feature1()
b1.feature2()
b1.feature3()
b1.feature4()   

c1=C()
c1.feature1()

e1=E()
e1.feature10()
```

```python
#Constructor in INheritance Python
#METHOD RESOLUTION ORDER(MRO)

class A:#SUPER CLASS/PARENT CLASS  #Grand father
    def __init__(self):
        print("in a INIT")
    def feature1(self):
        print("1")
    def feature2(self):
        print("2")
        

class B:#(A):#inheriting #SUBCLASS/CHILD CLASS #father
    def __init__(self):
        super().__init__()
        print("in B init")
    def feature1(self):
        print("22")
    def feature3(self):
        
        print("3")
    def feature4(self):
        print("4")
# multiple inheritence we use MRO
class C(A,B):
    def __init__(self):
        super().__init__()
        print("in C")
    def feat(self):#to represent super class we use super method
        super().feature2()
a1= C()
a1.feature1()
a1.feat()
```

```python
#here we are overloaidng that is changing th addition parameter to re add operators
class Students:
    def __init__(self,m1,m2):
        self.m1 = m1
        self.m2 = m2
    def __add__(self,other):
        m1= self.m1+other.m1
        m2= self.m1+other.m2
        s3 = Students(m1,m2)
        
        return s3
    def __gt__(self,other):
        r1 = self.m1+ self.m2
        r2 = other.m1+other.m2
        if r1>r2:
            return True
        else: 
            return False
    def __str__(self):
#         return self.m1,self.m2->returns the values in form of tuple
        return '{} {}'.format(self.m1,self.m2)#->returns in form of strings not in tuples type 
        
        
    
s1=Students(58,69)
s2=Students(69,65)
s3= s1+s2

#Prints the [MODULE CLASS OBJECT ADRESS]
#but after poly morphing the class it gives like this

print(s3.__str__())

# in order to efficiently print without the tuples like this we use
print(s3)
print(s1)
print(s2)
#for just printing 

print(s3.m1)

if s1>s2:
    print("s1 win")
else: 
    print("s2 wins")
```

```python
# 3)METHOD OVERLOADING

#CONCEPT: having the same method name and but having diffrent arguments to pass
#is called method overloading
# EX: def averge(a,b): def average(a,b,c):
class addition:
    def __init__(self,m1,m2):
        self.m1=m1
        self.m2=m2
    def sum(self,a=None,b=None,c=None):#to implement METHOD OVERLOADING TRIK 
        s= 0
        if a!=None and b!=None and c!=None:
            s = a+b+c
        elif a!=None and b!=None:
            s=a+b
        else:
            s=a
        return s

    
s1=addition(10,10)

print(s1.sum(3,))
```

```python
#Multi Threading in python->multi task [ex:MS WORD]
#To import the thrading we use threading

from threading import *
from time import sleep
class Hello(Thread):
    def run(self):
        for i in range(5):
            print("HELLO")
            sleep(1)#1 -> emplies 1 second 
    

class Hi(Thread):
    def run(self):
        for i in range(5):
            print("HI")
            sleep(1)

t1=Hello()
t2=Hi()

#in system we will have SHEDULERS which always allocate tie for getting output 
#if we chNGE FOR LOOP TO 500 times each all the vales execute and may colloide 
#we can slow down we can shedule the tasks 
t1.start()
sleep(0.2)
t2.start()

t1.join()
t2.join()

print("Bye")
```

```python
def sort(nums):
    for i in range(len(nums)-1,0,-1):
        for j in range(i):
            if nums[j]>nums[j+1]:
                temp = nums[j]
                nums[j]=nums[j+1]
                nums[j+1]=temp

nums =[5,3,8,6,7,2]
sort(nums)
print(nums)
```
